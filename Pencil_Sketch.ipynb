```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#let's look at the image first!
image=cv2.imread("dog.jpg")
cv2_imshow(image)
cv2.waitKey(0) #displays the image
```

```python
#converts image to greyscale
gray_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

```python
#The next step is to invert the above greycale image
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey(0)
```

```python
#we now blur the image using Guassian Function in OpenCV
blurred = cv2.GaussianBlur(inverted_image, (21,21), 0)
```

```python
#once we invert the blurred image, we can easily make it into a pencil sketch
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale = 245.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
